## L. Linear Regression Example

```python
##subset all independent variables  
import statsmodels.api as sm
#X = new_sample['YearBuilt']
X = new_sample[['YearBuilt', 'BedroomAbvGr']]
Y = new_sample['SalePrice']
X = sm.add_constant(X)

model = sm.OLS(Y, X, missing='drop').fit()
# compute with statsmodels, by adding intercept manually
import statsmodels.api as sm
X = sm.add_constant(X)
result = sm.OLS(y, X).fit()
print("pvalues\n",result.pvalues)
print("rsquared\n",result.rsquared)
print("coefficient parameters\n", result.params)
```

```python
from sklearn.linear_model import LinearRegression

# create linear regression object
mlr = LinearRegression()

# fit linear regression
mlr.fit(new_sample[['YearBuilt', 'BedroomAbvGr']], new_sample['SalePrice'])

# get the slope and intercept of the line best fit.
print(mlr.intercept_)
print(mlr.coef_)
```
